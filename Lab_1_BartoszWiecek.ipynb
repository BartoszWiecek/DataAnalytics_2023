### AD.2.

```python
import numpy as np
import numpy.random as rm

# 1000 samples of uniform distribution on interval [0, 1]
array_1 = rm.uniform(0, 1, 1000)

# 1000 samples of normal distribution with mean = 0 and standard deviation = 1.
array_2 = rm.normal(0, 1, 1000)

# 1000 samples of normal distribution with mean = 0 and standard deviation = 10.
array_3 = rm.normal(0, 10, 1000)

# 1000 samples of beta distribution for the following pairs of parameters (alpha, beta) = {(1, 1), (2, 2), (3, 9)}
array_4_1 = rm.beta(1, 1, 1000)
array_4_2 = rm.beta(2, 2, 1000)
array_4_3 = rm.beta(3, 9, 1000)

# 10000 samples of two-dimensional normal distributions each with mean = [[0], [0]] and covariance matrices E = {[[1, 0], [0, 1]], [[2, 1], [1, 2]]}
array_5_1 = rm.multivariate_normal([0, 0], [[1, 0], [0, 1]], 5000)
array_5_2 = rm.multivariate_normal([0, 0], [[2, 1], [1, 2]], 5000)
```

### AD.3.

```python
import matplotlib.pyplot as plt

titles = ['Uniform distribution', 
          r'Normal distribution ($\mu=0$, $\sigma=1$)', 
          r'Normal distribution ($\mu=0$, $\sigma=10$)', 
          'Beta distribution (\u03B1=1, \u03B2=1)',
          'Beta distribution (\u03B1=2, \u03B2=2)',
          'Beta distribution (\u03B1=3, \u03B2=9)']

arrays = [array_1, array_2, array_3, array_4_1, array_4_2, array_4_3]

for i, arr in enumerate(arrays):
    plt.hist(arr, 1000, density=True)
    plt.title(titles[i])
    plt.show()
```

### AD.4.

```python
titles = ['Two-dimensional normal distribution - array 1',
         'Two-dimensional normal distribution - array 2']

arrays = [array_5_1, array_5_2]

for i, arr in enumerate(arrays):
    plt.hist2d(arr[:, 0], arr[:, 1], bins=[50, 50])
    plt.title(titles[i])
    plt.show()
```

More bins = more smaller light points instead of less greather ones.

```python
import seaborn as sns
import pandas as pd

titles = ['Two-dimensional normal distribution - seaborn - array 1',
         'Two-dimensional normal distribution - seaborn - array 2']

arrays = [array_5_1, array_5_2]

for i, arr in enumerate(arrays):
    arr_df = pd.DataFrame(arr, columns = ['x', 'y'])
    sns.histplot(arr_df, x='x', y='y')
    plt.title(titles[i])
    plt.show()
```

### AD.5.

```python
from scipy.stats import halfnorm

y_points = np.zeros(1000)
alphas = np.zeros(1000)
betas = np.zeros(1000)
xses = np.zeros(1000)

for i in range(1000):
    alphas[i] = rm.normal(2, 0.1)
    betas[i] = rm.normal(0.2, 0.2)
    xses[i] = rm.normal(0, 3)
    
    sigma = abs(rm.normal(0, 1))
    micro = alphas[i] + (betas[i] * xses[i])

    y_points[i] = rm.normal(micro, sigma)
```

### AD.6. 

```python
x_array = np.linspace(-6, 6, 1000)

plt.scatter(xses, y_points)

for i in range(1000):
    y_tmp = [betas[i] * x_array[j] + alphas[i] for j in range(1000)]
    plt.plot(x_array, y_tmp)
    
plt.show()
```
